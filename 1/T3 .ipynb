### Ejercicio 2. 

Se tiene un sistema de ecuaciones diferenciales que modela un péndulo:


$\dfrac{d\theta}{dt} = v_{\theta}$


$\dfrac{dv_{\theta}}{dt} = -w^2_{0} sin\theta$





donde $w_{0}^2$ = w es una constante. Integre numéricamente y dibuje trayectorias solución para las condiciones
iniciales $v_{\theta}$ = 0 y $\theta_{0}= \frac{n\pi}{10}$ con n = 1,2,3,..., 9.

```python
def ec_pend(ini,t):
    theta = ini[int((9*pi)/10)]
    v0 = ini[0]
    w=2
    f1= v0
    f2 = -w*sin(theta)
    return array ([theta,v0,f1,f2])
   
        
    

```

```python

t = linspace(0,5)
c_ini = [0.,0.,10.,10.]
sol = odeint(ec_pend,c_ini,t)


x,y = sol[:,0],sol[:,1]

plot(x,y,label="n=9")
gca().set_aspect('equal')
xlim([0,60]),ylim([0,15]),legend()
show()


```

### Ejercicio 3. 

Se tiene el siguiente sistema de ecuaciones que modela un oscilador armónico amortiguado:

$\dfrac{dx}{dt} = v_{x}$


$\dfrac{dv_{x}}{dt}$= $-w^2_{0}x-2\gamma w_{0}v_{x}$



a) Encuentre las soluciones del sistema anterior con $\gamma <1$ y dibuje la solucion en el plano (x,$v_{x}$) para varias condiciones iniciales de dichas variables, por lo menos 10 que barran un conjunto que represente bien a las soluciones. 

```python
def osc_arm(v,t):
    x= v[0]
    v_x = v[1]
    dx = v_x
    w=2
    gamma=-1
    dv_x = int(((-(w**2))*((x)))-(2*(gamma*w*v_x)))
    return [dx,dv_x]
```

```python
p = 300
t = arange(0,p,1) # variable tiempo 
figure(figsize=(7,7))
sol = odeint(osc_arm,[0,1],t) #vamos a integrar la funcion desde la condicion inicial hasta el tiempo t
plot(sol[:,0], sol[:,1],"-", color='y', label="$v_0=(0,1)$") #graficar las entradas en la columna 0 vs la columna 1
xlabel(' $x$ ')
ylabel(' $v_x$ ')
title(u'$(x,v_x)$',fontsize=10)
legend(bbox_to_anchor=(1, 1), loc=2,borderaxespad=0.),grid(True)
minorticks_on()
show()
```

### Ejercicio 4.

$\dot{v}_{x} = v_{y}B_{0}$

$\dot{v}_{y} = -v_{x}B_{0}$

con $B_{0} = 2$ y condiciones iniciales $\vec{v}_{0}= (0,1).$


a) Como es la trayectoria en el plano ($v_{x}$, $v_{y}$)




```python
def sis_1(v,t,B=2): #Definimos la funcion sis_1 con los argumentos v, t=tiempo y el campo 
    v_x = v[0]
    v_y = v[1]
    dv_x = v_y*B
    dv_y = -v_x*B
    return [dv_x,dv_y]
```

```python
p = 300
t = arange(0,p,1) # variable tiempo 
figure(figsize=(7,7))
sol = odeint(sis_1,[0,1],t) #vamos a integrar la funcion desde la condicion inicial hasta el tiempo t
plot(sol[:,0], sol[:,1],"-", color='y', label="$v_0=(0,1)$") #graficar las entradas en la columna 0 vs la columna 1
xlabel(' $v_x$ ')
ylabel(' $v_y$ ')
title(u'$(v_x,v_y)$',fontsize=10)
legend(bbox_to_anchor=(1, 1), loc=2,borderaxespad=0.),grid(True)
minorticks_on()
show()
```

```python
cuadro = figure() # donde realizaremos la animacion
ejes = axes(xlim=(-2, 2), ylim=(-2, 2))
gca().set_aspect('equal', adjustable='box')
x,y = sol[:,0],sol[:,1] #x y y son soluciones en la columna cero y primer columna al integrar la funcion sis_1
plot(x,y, 'bo', markersize=1.0)
trayectoria, = ejes.plot([], [],'b-',markersize=30.0)
particula, = ejes.plot([], [], 'y.', markersize=15.0)

def init():
    trayectoria, = ejes.plot([], [])
    particula.set_data([],[])
    return trayectoria, particula

def animate(i):
    trayectoria.set_data(x[:i], y[:i])
    particula.set_data(x[i],y[i])
    return trayectoria, particula

animation.FuncAnimation(cuadro, animate, init_func=init, frames=50, interval=80, blit=True)
```
