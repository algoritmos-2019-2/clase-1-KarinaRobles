#### Ejercicio 4

Realice una función que se llame mi_sinc(x,N_max=10), donde $x$ es un número real y $N\_ max$ es un entero que por default es 10. Esta función calcula la siguiente serie
$$
\sum ^{N_{max}} _{n=0} = \frac{(-1)^n}{(2n+1)!} x^{2n} = 1-\frac{x^2}{3!} + \frac{x^4}{5!} - ...
$$


A. Haga la gráfica de la serie anterior en el intervalo $x \epsilon [-5\pi ,5\pi ]$ con el valor de $N_{max}$ que viene por default.


```python
def mi_sinc(x,N_max=10): 
    serie = 0 #serie empieza en 0
    for n in range(0,N_max): #n corre de 0 a N_max
        serie += ((-1)**n)*(x**(2*n))/(factorial(2*n+1)) #serie va sumando para cada n 
    return serie 
```

```python
xint = linspace(-5*pi,5*pi,300) 
yint = [] #yint es una lista vacia
for i in range(len(xint)): #por cada elemento de xint
    y = mi_sinc(xint[i]) 
    yint.append(y) #appendo y a yint

plot(xint,yint) 
title('Serie de la funcion Sinc(x) a 10 orden',fontsize=20)
xlabel('$x$',fontsize=16)
ylabel('$y$',fontsize=16)
ylim(-4,2)
```

B. Ahora haga la gráfica de la función anterior en el mismo intervalo para $x$ pero ahora con $N_{max} = 0,2,4,6,...,20$ y observe la evolución de la serie conforme se aumenta $N_{max}$. (Todas las gráficas de este inciso deben estar en la misma gráfica).

```python
N = arange(0,22,2) # N es un arange de 0 a 20 en pasos de 2
xint = linspace(-5*pi,5*pi,300)
for n in N: 
    yint = [] 
    for i in range(len(xint)): 
        y = mi_sinc(xint[i],n) #y es igual a la serie sinc correpondiente a la n que le toca
        yint.append(y) #appendo y a yint
    plot(xint,yint,label="N="+str(n)) 
    title('Serie de taylor de la funcion Sinc(x)',fontsize=20) 
    xlabel('$x$',fontsize=16)
    ylabel('$y$',fontsize=16),legend(loc=5)
    xlim(-19,28),ylim(-4,2)
```

```python
def tir_v2 (x_v,t,g=9.8, gamma= 0.1): 
    #x_v ) x,y,vx,vy
    dx=x_v[2]
    dy=x_v[3]
    v=sqrt(x_v[2]**2. + x_v[3]**2. )
    dvx= -v*x_v[2]
    dvy =-g-v*x_v[3]
    return dx,dy,dvx,dvy
```
